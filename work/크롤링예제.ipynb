### 1. melon 차트 1-50위 가수명과 제목 출력하기
#### https://www.melon.com/chart/index.htm

```python

from bs4 import BeautifulSoup as soup
import urllib.request 
import requests 


hdr = {
        'accept-language': 'ko,th;q=0.9,en-US;q=0.8,en;q=0.7,ru;q=0.6,ko-KR;q=0.5,la;q=0.4',
        "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36"
    }

url = "https://www.melon.com/chart/index.htm"
html = requests.get(url, headers=hdr).text
bs_obj = soup(html,"html.parser")

title = bs_obj.select("div.ellipsis.rank01 > span > a")
name =  bs_obj.select("div.ellipsis.rank02 > a")
i = 1

for t, n in zip(title, name) :
        print(i, '등 제목:', t.string, '/', '가수:', n.string)
        i = i+ 1
        if i == 51:
            break
            

```

### 2. 베스트셀러 순위와 제목 출력

```python


from bs4 import BeautifulSoup
import urllib.request

response=urllib.request.urlopen("https://search.shopping.naver.com/book/search?bookTabType=BEST_SELLER&catId=50005542&pageIndex=1&pageSize=40&query=%EB%B2%A0%EC%8A%A4%ED%8A%B8%20%EC%85%80%EB%9F%AC&sort=REL")
html_str=response.read().decode("utf-8")
#print(html_str)
bs_obj=BeautifulSoup(html_str,"html.parser") 
a=bs_obj.select(".bookListItem_text__bglOw > span >span") #순번
b=bs_obj.select(".bookListItem_text__bglOw > span:nth-child(2)") #타이틀


for a1,b1 in zip(a,b):
    print(a1.string,b1.string)
```

### 3. 쿠팡 상품명과 가격 출력

```python
from bs4 import BeautifulSoup as soup
import urllib.request 
import requests #쿠팡은 이걸로만 가능

#쿠팡 크롤링 헤더 정보 : https://hacktam.kr/etclec/244

hdr = {
        'accept-language': 'ko,th;q=0.9,en-US;q=0.8,en;q=0.7,ru;q=0.6,ko-KR;q=0.5,la;q=0.4',
        "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36"
    }

url = "https://www.coupang.com/np/search?q=%EC%9D%B8%EA%B0%84%EC%82%AC%EB%A3%8C&channel=user&component=&eventCategory=SRP&trcid=&traid=&sorter=scoreDesc&minPrice=&maxPrice=&priceRange=&filterType=&listSize=50&filter=&isPriceRange=false&brand=&offerCondition=&rating=0&page=1&rocketAll=false&searchIndexingToken=1=9&backgroundColor="
html = requests.get(url, headers=hdr).text
bs_obj = soup(html,"html.parser")
name = bs_obj.select("div.name ")
price = bs_obj.select("em.sale > strong.price-value")

for name, price in zip(name, price):
    print(name.string, price.string)
```

### 4. 커피빈 매장 수집
#### https://www.coffeebeankorea.com/store/store.asp

```python
#교재의 방법은 문제가 많아서 다른방법으로 시도
import requests 
import json
import re #정규식 처리 모듈

url_head='https://www.coffeebeankorea.com/store/store_data2.asp?lat=&lng=&chk1=0&chk2=0&chk3=0&chk4=0&chk5=0&chk6=0&chk7=0&chk8=0&chk9=0&chk10=0&chk11=0&chk12=0&keyword=&StoreLocal='
url_tail='&StoreLocal2=&storeNo='
store_locals=('서울','경기','인천','강원','경남','광주','대구','대전','부산','제주','충남','충북')
sum=0
for local in store_locals:
    response = requests.get(url_head+local+url_tail) 

    #re.sub(정규표현식, 치환문자, 대상문자열) -> 키값에 ' '붙여줌 (json 파싱 위해 )
    single_quotes_json=re.sub('([a-zA-Z0-9]+) :', r"'\1' :", response.text) 
  
    #'를 "변경 (json 파싱 위해 )
    json_data=single_quotes_json.replace("'",'"')
    local_data=json.loads(json_data)  
    for store in local_data:
        print(store['StoreName'],store['StoreOpendate'])
        sum+=1
print("매장총갯수",sum)
```

### 6. 정규식

```python
#정규식(regular expression)
import re #정규식사용하기 위한 모듈

temp_str='ython abthon aa'
# ^ : 문자열의 시작을 나타낸다.
# . : 글자 하나를 나타낸다.
m=re.match('..thon',temp_str) #^생략해도 ^삽입해서 찾는다.(첫단어만 확인)
print(m)

#match는 필요없다.
m=re.search('..thon',temp_str) #우리가 생각그거,즉 문자열 전체를 검색(첫단어만 검색할려면 앞에 ^붙인다)
print(m)

m=re.search('ca.e','Good care caae')
print("입력받은 문자열", m.string)
print("일치하는 문자열", m.group())
print("일치하는 문자열의 시작인덱스", m.start())
print("일치하는 문자열의 끝인덱스", m.end())
print("일치하는 문자열의 시작과끝 인덱스", m.span())


'''
. :문자하나              ex>app.e -> apple
^ :문자열의 시작
$ :문자열의 끝
? :문자하나(없어도 된다.) ex> app?e ->apple, appe 
| :or(다자택일)          ex> a|bpple -> apple, bpple
반복기호: * 0번이상, + 1번이상, {n~m} n~m번 반복 ex>a*pple -> apple,aaaaapple,abpple(X),pple

문자클래스[]: []사이의 문자들과 매치 ex>[abc]pple ->apple,bpple,cpple,abpple(X)
그룹 ():()사이의 문자와 모두매치   ex> (abc)pple -> abcpple

문자클래스 특수용도
[0-9] 모든숫자를 의미
[a-z] a~z까지를 의미
[A-Z] A~Z까지를 의미
[a-zA-Z] 모든 알파벳
[ㄱ-ㅎㅏ-ㅣ가-힣]  ex>ㅋㅋㅋ,ㅠㅠㅠ,한글매칭
\d : 모든숫자
\D : (반대) 숫자가 아닌거 -> [^0-9]
\w : 모든문자(문자+숫자)
\W : (반대) 문자가 아닌거-> [^a-zA-Z0-9]
\s : 공백(모든공백) \t(탭) \n(줄바꿈) \r(캐리지리턴) 
\S : 공백이 아닌거
\b : 단어의 경계, 공백, 탭, 컴마, 대시등
\B : 단어의 경계가 아닌것

- re 모듈에서 제공하는 메소드들
mathch() : 문자열의 처음부터 정규식과 매칭되는지 조사
search() : 문자열의 전체를 검색해서 정규식과 매칭되는지 조사
findall() : 정규식과 매치되는 모든 문자열을 리스트로 돌려준다.
split(): 정규식과 매치되는 문자열을 기준으로 파싱하여 리스트로 돌려준다.
'''
```

```python
mlist=re.findall('[\s]*berry','berry apple, strawberry kiwi ,straw berry banana, rasberry')
print(mlist)

#그룹캡처
m = re.search('\d{4}-(\d\d)-(\d\d)','2021-03-09')
print(m.group())
print(m.group(0))
print(m.group(1))
print(m.group(2))
```

```python
tr_tag = '<tr hef="dssdd", id="abc123", class="ddd">hello</tr>'
m = re.search('<tr.*id="(.*)",.*>(.*)</tr>', tr_tag)
print(m.group(0)) #그룹 전체 매치 값
print(m.group(1)) #tr 태그 내의 속성 id 값
print(m.group(2)) #tr 태그의 value 값
```
